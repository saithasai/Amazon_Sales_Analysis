```python
data["Year"] = data["Order Date"].dt.year
data.groupby('Year')['Total Profit'].mean().plot()
plt.xlabel('Order Date')
plt.ylabel('Total Profit')
plt.title('Profit per year')
```

```python
names = data.Country.value_counts().index
value = data.Country.value_counts().values
plt.pie(value[:20],labels=names[:20],autopct='%1.2f%%',radius=1.8)
```

**Most sales occurrences happened in the country 'The Gambia'**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
data.groupby('Region')['Order Date'].count().plot(kind='bar', color=['green', 'orange', 'blue', 'red', 'purple', 'pink', 'yellow', 'brown', 'aqua', 'grey'], ax=ax1)
ax1.set_ylabel('Count')
ax1.set_title('Count of occurrences of Item Type')

pd.pivot_table(data, values='Order Date', index='Region', aggfunc='count').plot(kind='bar', ax=ax2)
ax2.set_ylabel('Count')
ax2.set_title('Count of occurrences of Item Type')

plt.tight_layout(pad=5)
plt.show()
```

**Most received orders are from Sub-Saharan Africa and least from North America**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
data.groupby('Item Type')['Total Revenue'].sum().plot(kind='bar', color=['green', 'orange', 'blue', 'red', 'purple', 'pink', 'yellow', 'brown', 'aqua', 'grey'], ax=ax1)
ax1.set_ylabel('Sum')
ax1.set_title('Total Revenue by Item Type')

pd.pivot_table(data, values='Total Revenue', index='Item Type', aggfunc='sum').plot(kind='bar', ax=ax2)
ax2.set_ylabel('Sum')
ax2.set_title('Total Revenue by Item Type')

plt.tight_layout(pad=5)
plt.show()
```

**Maximum total revenue generated from the item Cosmetics   
Minimum total revenue generated from the item Fruits**

```python
fig,(ax1,ax2)=plt.subplots(1, 2, figsize=(12, 6))
data.groupby('Order Priority')['Total Revenue'].sum().plot(kind='bar',color=['red','aqua','brown','black'],ax=ax1)
ax1.set_ylabel('Total Revenue')

data.groupby('Order Priority')['Total Profit'].sum().plot(kind='bar',color=['red','aqua','brown','black'],ax=ax2)
plt.ylabel('Total Profit')

plt.tight_layout(pad=5)
plt.show()
```

**High priority tasks yielded more revenue and Critical tasks yielded least revenue    
High priority tasks yielded more profit and Critical tasks yielded least profit**
